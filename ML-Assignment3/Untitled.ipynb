```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def rotate180(img):
    p, q = img.shape
    img2 = np.zeros(img.shape)
    for i in range(p):
        for j in range(q):
            img2[i, q-j-1] = img[i, j]
    return img2
```

```python
from matplotlib import pyplot as plt
import numpy as np
image = np.zeros((100, 200))
for i in range(100):
    for j in range(200//2):
        image[i, j] = 1
plt.imshow(image, cmap='gray')
```

```python
output = rotate180(image)
plt.imshow(output, cmap='gray')
```
