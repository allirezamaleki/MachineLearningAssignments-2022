```python
import numpy as np
```

```python
def gradient_descent(x, y, iterations):
    w = 0
    b = 0
    m = len(x)
    learning_rate = 0.1
    
    for i in range(iterations):
        y_pred = w * x + b
        J = 1/(2 * m) * np.sum(np.dot(y_pred - y, y_pred - y))
        
        w = w - learning_rate * (1/m * np.sum(np.dot(y_pred - y, x)))
        b = b - learning_rate * (1/m * np.sum(y_pred - y))
        
        print(f'iteration {i+1}: w = {w}, b = {b}, cost = {J}')
        
    return w, b,  J
```

```python
# daraset
x = np.array([1, 2, 3])
y = np.array([1, 2, 3])
```

```python
gradient_descent(x, y, 10)
```

```python
gradient_descent(x, y, 100)
```
