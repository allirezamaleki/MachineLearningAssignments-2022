## Dataset (Traditionaly)

```python
dataset = pd.read_csv('./iris.data')
print(f'Current Column names: {dataset.columns}')

new_row = {'sepal length': 5.1, 'sepal width': 3.5, 'petal length': 1.4, 'petal width': 0.2, 'class': 'Iris-setosa'}
new_columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class']

dataset.columns = new_columns
dataset = pd.concat([pd.DataFrame([new_row]), dataset], ignore_index=True)
```

```python
dataset.head()
```

```python
len(dataset)
```

```python
label_encoder = LabelEncoder()
dataset['class'] = label_encoder.fit_transform(dataset['class'])
```

```python
dataset.head()
```

```python
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values

x[:1], y[:1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 42)

print(f'''length of All records: {len(x)}
length of train datasets: {len(x_train)}
length of test datasets: {len(x_test)}''')
```

```python
# Feature Scaling
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
print(f'Sample scaled x_train: {x_train[0]},\nSample scaled x_test: {x_test[0]}')
```

```python
RFC = RandomForestClassifier(n_estimators = 10)
RFC.fit(x_train, y_train)

y_pred = RFC.predict(x_test)
print(f'Random Forest Classifer Accuracy: {np.round(accuracy_score(y_test, y_pred) * 100, 2)}%')
```

```python
DTC = DecisionTreeClassifier(criterion = 'gini')
DTC.fit(x_train, y_train)

y_pred = DTC.predict(x_test)
print(f'Decision Tree Classifer Accuracy: {np.round(accuracy_score(y_test, y_pred) * 100, 2)}%')
```

```python
naiveBayes = GaussianNB()
naiveBayes.fit(x_train, y_train)

y_pred = naiveBayes.predict(x_test)
print(f'Naive Bayes Classifer Accuracy: {np.round(accuracy_score(y_test, y_pred) * 100, 2)}%')
```

```python
NN = MLPClassifier(hidden_layer_sizes=(7, 7, 7), max_iter = 350)
NN.fit(x_train, y_train)

y_pred = NN.predict(x_test)
print(f'Neural Network Classifer Accuracy: {np.round(accuracy_score(y_test, y_pred) * 100, 2)}%')
```
